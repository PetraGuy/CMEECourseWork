The task is to get linear models for log.pred.mass ~ log.prey.mass by Feeding type, by predator lifestage - in other words, by factors in two columns. To do this I'll use a smimple function to start with - mean, and I'll create a simple, smaller dataframe. First with just one factor

```r
x = seq(1,100, by=1) ; x
```

```r
A = rep(1:5, each=20);A
```

```r
df = data.frame(cbind(x,A))
```

```r
means = tapply(df$x,df$A, mean); means
```

What if there are two columns of factors

```r
B = rep(1:4, each=25);B
```

```r
df = data.frame(cbind(x,A,B)); df
```

```r
aggregate(x ~ A + B, data = df, FUN= "mean" )
```
